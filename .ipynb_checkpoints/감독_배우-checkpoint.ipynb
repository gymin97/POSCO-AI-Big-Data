```python
import pandas as pd
from ast import literal_eval
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.metrics.pairwise import linear_kernel, cosine_similarity
```

```python
inventory = pd.read_csv('movie_inventory1.csv', engine = 'python')
meta = pd.read_csv('movie_meta.csv', engine = 'python')

inventory.drop('Unnamed: 0', axis = 1, inplace = True)

display(inventory)
display(meta)
```

```python
meta1 = meta[['movie_id', 'description']]
data = pd.merge(inventory, meta1, on = 'movie_id')
data
```

```python
director_list = list(data['director'])
actor_list = list(data['actor_1'])
genre_list = list(data['Genre_1'])
soup_list = []

for i in range(len(director_list)):
    soup = director_list[i].replace(" ", "") + str(" ") + \
    actor_list[i].replace(" ", "") + str(" ") + genre_list[i]
    soup_list.append(soup)

soup_list
```

```python
data['soup'] = soup_list
data.head()
```

```python
count = CountVectorizer(analyzer='word', ngram_range=(1, 2), min_df=0,
                       stop_words='english')
count_matrix = count.fit_transform(data['soup'])
```

```python
print(count.vocabulary_)
```

```python
cosine_sim = cosine_similarity(count_matrix, count_matrix)
```

```python
print(cosine_sim)
```

```python
titles = data['title']
indices = pd.Series(data.index, index = data['title'])
```

```python
def getrecommandations(title):
    index = indices[title]
    sim_scores = list(enumerate(cosine_sim[index]))
    sim_scores = sorted(sim_scores, key=lambda x:x[1], reverse=True)
    sim_scores = sim_scores[1:31]
    movie_indices = [i[0] for i in sim_scores] 
    return titles.iloc[movie_indices] 
```

```python
getrecommandations('Harry Potter and the Goblet of Fire')
```
